```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
num_of_patients=[32, 28, 35, 30, 29, 27, 31, 34, 33, 30]
num_of_patients=pd.DataFrame(num_of_patients)
```

```python
num_of_patients.mean()[0]
```

```python
num_of_patients.median()[0]
```

```python
num_of_patients.mode()
```

```python
q1=num_of_patients.quantile(0.25)
q3=num_of_patients.quantile(0.75)
iqr=q3-q1
lower_fence=q1-1.5*iqr
upper_fence=q3+1.5*iqr
daily_patients=num_of_patients.iloc[:,0]
outliers=daily_patients[(daily_patients<lower_fence[0])|(daily_patients>upper_fence[0])]
len(outliers)
```

```python
plt.figure(figsize=(7,7))
sns.boxplot(data=num_of_patients)
plt.title('Boxplot of Daily number of patients')
plt.show()
```

* as there are no outliers present hence mean() value best represents the patients admission
* if there would be outliers present in daily number of patients admissions then median or mode would be the best representation of the daily number of patient admissions

```python
increased_patients=list(map(lambda x:round(0.1*x)+x,num_of_patients[0]))
increased_patients=pd.DataFrame(increased_patients)
```

```python
increased_patients.mean()[0]
```

```python
((increased_patients.mean()[0]-num_of_patients.mean()[0])/num_of_patients.mean()[0])*100
```

```python
(increased_patients.median()[0]-num_of_patients.median()[0])/num_of_patients.median()[0]*100
```

```python
(increased_patients.mode()[0][0]-num_of_patients.mode()[0][0])/num_of_patients.mode()[0][0]*100
```

* all measures (central tendencies: mean,median,mode )also increased by approximately 10%
* relative relations remains unchanged
